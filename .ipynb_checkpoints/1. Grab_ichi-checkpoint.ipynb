```python
#Clear data folder
def clean():
    filelist = [ f for f in os.listdir(path) if f.endswith(".csv") ]
    for f in filelist:
        os.remove(os.path.join(path, f))

    filelist = [ f for f in os.listdir(path) if f.endswith(".png") ]
    for f in filelist:
        os.remove(os.path.join(path, f))

def countdown(t=60, step=1,msg='Sleeping'):  # in seconds
    pad_str = ' ' * len('%d' % step)
    for i in range(t, 0, -step):
        sys.stdout.write('%s for %d seconds   %s\r' % (msg, i, pad_str)),
        sys.stdout.flush()
        time.sleep(step)
        
def get_filename(market, tick_interval):
    return '%s_%s.csv' %  (market, tick_interval)
def get_filenamed(market, tick_interval):
    return '%s_%s_add.csv' %  (market, tick_interval)
    
def getData(market, tick_interval):
    url = 'https://api.binance.com/api/v3/klines?symbol='+market+'&interval='+tick_interval
    response = requests.get(url).json()
    data={
            'OpenTime':[x[0]for x in response],
            'Open':[x[1]for x in response],
            'High':[x[2]for x in response],
            'Low':[x[3]for x in response],
            'Close':[x[4]for x in response],
            'Volume':[x[5]for x in response]
        }

    df = pd.DataFrame(data)
    df['Open']= df['Open'].astype(float)
    df['High']= df['High'].astype(float)
    df['Low']= df['Low'].astype(float)
    df['Close']= df['Close'].astype(float)
    val=3600000 #1h to ms
    last_index = df.iloc[-1:].index[0]
    last_date = df.iloc[-1,df.columns.get_loc("OpenTime")]

    if tick_interval == '1h':
        interval = 1
    elif tick_interval == '4h':
        interval = 4
    elif tick_interval == '12h':
        interval = 12
    elif tick_interval == '1d':
        interval = 24
    elif tick_interval == '3d':
        interval = 3*24
    elif tick_interval == '1w':
        interval = 7*24

    for i in range(1,31):
        df.loc[last_index+1+i, 'OpenTime'] = last_date + i*val*interval
    return df

def ichimoku(df):
    tenkan_window = 20
    kijun_window = 60
    senkou_span_b_window = 120
    cloud_displacement = 30
    chikou_shift = -30
    # Tenkan 
    tenkan_sen_high = df['High'].rolling( window=tenkan_window ).max()
    tenkan_sen_low = df['Low'].rolling( window=tenkan_window ).min()
    tenkan = df['tenkan_sen'] = (tenkan_sen_high + tenkan_sen_low) /2
    # Kijun 
    kijun_sen_high = df['High'].rolling( window=kijun_window ).max()
    kijun_sen_low = df['Low'].rolling( window=kijun_window ).min()
    kijun = df['kijun_sen'] = (kijun_sen_high + kijun_sen_low) / 2
    # Senkou Span A 
    sen_a =df['senkou_span_a'] = ((df['tenkan_sen'] + df['kijun_sen']) / 2).shift(cloud_displacement)
    # Senkou Span B 
    senkou_span_b_high = df['High'].rolling( window=senkou_span_b_window ).max()
    senkou_span_b_low = df['Low'].rolling( window=senkou_span_b_window ).min()
    sen_b =df['senkou_span_b'] = ((senkou_span_b_high + senkou_span_b_low) / 2).shift(cloud_displacement)
    # Chikou
    chikou=df['chikou_span'] = df['Close'].shift(chikou_shift)
    return df
 
def run():
    df = getData(x, y)
    df.to_csv(filename,index = None, header=True)
    df.to_csv(filenamed)
    df = ichimoku(df)
    df['OpenTime'] = pd.to_datetime(df['OpenTime'], unit='ms')
    #df = df.fillna(0)
    df.to_csv(filenamed,index = None, header=True)
    os.remove(filename)
    print("Grabbed:",x, y, filenamed) 
```
